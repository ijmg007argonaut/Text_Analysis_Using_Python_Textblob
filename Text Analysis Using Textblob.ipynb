```python
# SUMMARY:
# Sentiment text analysis was performed for 40 sentences using pythonâ€™s textblob library. For each
# sentence, words, word tags, sentiment, and polarities were extracted. Sentiment results from the textblob
# library were composed of named tuplets containing both polarity and subjectivity. Polarity is a float value
# within the range of [-1.0 to +1.0] where -1 indicates a very negative sentiment, +1 represents a very positive
# sentiment, and 0 indicates a neutral sentiment. Subjectivity is a float value within the range [0.0 to +1.0]
# where 1.0 is very subjective (expresses personal feelings, views, opinions, allegations, concerns, beliefs, and
# speculations) and 0.0 is very objective (expresses simple facts). A scatter plot was generated to allow
# comparison between the subjectivity and polarity of each of the 40 sentences.
```

```python
# Import Modules
from textblob import TextBlob
import numpy as np
import matplotlib.pyplot as plt
import os  # operating package to build directory

# setting working directory
directory = "/media/ijmg/SSD_FOUR_TB/ACADEMICS_101/a_Florida Atlantic University/MSAI/CAP 6776 - Info Retrieval/Assignment3/"
os.chdir(directory)

# import data file
the_file = ("assignment_3_data.txt")
```

```python
# Function to find text words
def find_words (file):
    print("\n---SENTENCE WORDS---\n")
    count = 1
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        print("SENTENCE #",count ,"WORDS:",TextBlob(line).words)
        count = count + 1
    text_file.close

find_words(the_file)
```

```python
# Function to find text tags
def find_tags (file):
    print("\n---SENTENCE TAGS---\n")
    count = 1
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        print("SENTENCE #",count ,"TAGS:",TextBlob(line).tags)
        count = count + 1
    text_file.close

find_tags(the_file)
```

```python
# Function to find text sentence sentiments
def find_sentiment (file):
    print("\n---SENTENCE SENTIMENTS---\n")
    count = 1
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        print("SENTENCE #",count ,"SENTIMENT:",TextBlob(line).sentiment)
        count = count + 1
    text_file.close

find_sentiment(the_file)
```

```python
# Function to find text sentence polarities
def find_polarity (file):
    print("\n---SENTENCE POLARITIES--\n")
    count = 1
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        print("SENTENCE #",count ,"POLARITY:",TextBlob(line).sentiment.polarity)
        count = count + 1
    text_file.close

find_polarity(the_file)
```

```python
# Function to plot polarities
def plot_polarity (file):
    polarity_array = np.array([], dtype = np.float32)
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        the_polarity = TextBlob(line).polarity
        polarity_array = np.append(polarity_array, the_polarity)
    text_file.close
    x_polarity = np.linspace(1,len(polarity_array), len(polarity_array))
    y_polarity = polarity_array
    return x_polarity, y_polarity

# Function to plot subjectivities
def plot_subjectivity (file):
    subjectivity_array = np.array([], dtype = np.float32)
    text_file = open(file, 'r')
    the_lines = text_file.readlines()
    for line in the_lines:
        the_subjectivity = TextBlob(line).subjectivity
        subjectivity_array = np.append(subjectivity_array, the_subjectivity)
    text_file.close
    x_subjectivity = np.linspace(1,len(subjectivity_array), len(subjectivity_array))
    y_subjectivity = subjectivity_array
    return x_subjectivity, y_subjectivity

x_polarity, y_polarity = plot_polarity(the_file)
plt.scatter(x_polarity, y_polarity, color = 'blue', label = 'Sentence Polarities')
x_subjectivity, y_subjectivity = plot_subjectivity(the_file)
plt.scatter(x_subjectivity, y_subjectivity, color = 'red', label = 'Sentence Subjectivities')
# Display polarities and subjectivities
plt.title("Sentiment Analysis: Sentence Polarity and Subjectivity Visualization")
plt.xlabel("Arbitrary sentence number ... sentence #1 ... sentence #2 ... etc to 40")
plt.ylabel("Sentence Polarity (in blue) Subjectivity (in red)")
plt.xlim(0,41)
for i in range(1,len(x_subjectivity)+1):
    plt.axvline(x=i, color=	'#838B8B', linestyle='--')
plt.legend()
plt.show()
```
